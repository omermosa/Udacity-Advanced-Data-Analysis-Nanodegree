```python
#violin plot
sb.violinplot(data=new_df,x='user_type',y='duration_sec',color=sb.color_palette()[9],inner='quartile');
```

```python
#box plot
sb.boxplot(data=new_df,x='user_type',y='duration_sec',color=sb.color_palette()[9]);
```

Customer members tend to spend more time in the ride- they have higher mean and ranges. Next, Let's see how bike sharing affect the duration

```python
#violin plot
sb.violinplot(data=new_df,x='bike_share_for_all_trip',y='duration_sec',color=sb.color_palette()[9],inner='quartile');
```

```python
sb.boxplot(data=new_df,x='bike_share_for_all_trip',y='duration_sec',color=sb.color_palette()[9]);
```

Surprisingly, when there is no bike sharing, the duraiton tends to be slightly higher. Next, let's see the birth year relation with duration.

Since most of the data lies in birth years > 1950, let's look ar the plot there. The plot doesn't show any particular trend. Next let's see the distance.

```python
sb.regplot(data=new_df,x='dist',y='duration_sec',scatter_kws={'alpha':0.2},fit_reg=False)
```

the plot shows nothing intersting except that most of  the data is clustered in the region of distance less than 10. There is a slight upward trend between the distance and the duration, but not a thing that can be considered a direct relation.

## Multivariate Exploration

> Create plots of three or more variables to investigate your data even
further. Make sure that your investigations are justified, and follow from

```python
def facet_grid(cat_fet,x,y,x_lim=None,y_lim=None):
    g=sb.FacetGrid(data=df,col=cat_fet,xlim=x_lim)
    g.map(plt.scatter,x,y,alpha=0.2)
   
```

### distance vs duration split by Categorical variables

```python
# split by gender and observe distance with duration
facet_grid('member_gender','dist','duration_sec',[0,20])
```

```python
# split by user type and observe distance with duration

facet_grid('user_type','dist','duration_sec',[0,15])
```

```python
# split by bike sharing and observe distance with duration

facet_grid('bike_share_for_all_trip','dist','duration_sec',[0,15])
```

I plotted The three categorical variable to split the data of the plots between the distance and the duration. It seems that there is no relation. all subplots appear to have similar ranges in x and y axes.
